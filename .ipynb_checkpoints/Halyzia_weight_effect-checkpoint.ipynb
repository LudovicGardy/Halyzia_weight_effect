# 1. Fast Ripples

```python
### Base model

import pandas as pd
import numpy as np

path_list = []

path_list.append(r"/Users/gardy/Documents/GardyL/Python/Halyzia_weight_effect/weight-1base_detection_summary_20022023_12-57-46.xlsx")
path_list.append(r"/Users/gardy/Documents/GardyL/Python/Halyzia_weight_effect/weight-1_detection_summary_20022023_15-25-00.xlsx")
path_list.append(r"/Users/gardy/Documents/GardyL/Python/Halyzia_weight_effect/weight-2_detection_summary_20022023_15-02-13.xlsx")
path_list.append(r"/Users/gardy/Documents/GardyL/Python/Halyzia_weight_effect/weight-5_detection_summary_20022023_11-28-06.xlsx")
path_list.append(r"/Users/gardy/Documents/GardyL/Python/Halyzia_weight_effect/weight-10_detection_summary_20022023_11-45-32.xlsx")
                 
step1_db0 = []
step1_db5 = []
step1_db10 = []
step1_db15 = []

for path in path_list:       
 
    print("")
    print(path)
    df = pd.read_excel(path)
    print(df.keys())
    print("")

    nb_pass_step1 = {"0dB": 0, "5dB": 0, "10dB":0, "15dB":0}
    nb_pass_step2 = {"0dB": 0, "5dB": 0, "10dB":0, "15dB":0}
    nb_pass_both = {"0dB": 0, "5dB": 0, "10dB":0, "15dB":0}

    n_lines = df.shape[0]

    for i in range(n_lines):
        p_step1 = False
        p_step1 = False
        dB = df["chan_name"][i].split("-")[1].replace(' SNR ', '').replace('dB ', "dB")

        if df["pass_step1"][i]:
            p_step1 = True
            nb_pass_step1[dB]+=1

        if df["pass_step2"][i]:
            p_step2 = True
            nb_pass_step2[dB]+=1

        if p_step1 and p_step2:
            nb_pass_both[dB]+=1
            
    step1_db0.append(nb_pass_step1["0dB"])
    step1_db5.append(nb_pass_step1["5dB"])
    step1_db10.append(nb_pass_step1["10dB"])
    step1_db15.append(nb_pass_step1["15dB"])

    print("0dB")
    print("------")
    print(f"Pass step 1: {nb_pass_step1['0dB']}")
    print(f"Pass step 2: {nb_pass_step2['0dB']}")
    print("")

    print("5dB")
    print("------")
    print(f"Pass step 1: {nb_pass_step1['5dB']}")
    print(f"Pass step 2: {nb_pass_step2['5dB']}")
    print("")

    print("10dB")
    print("------")
    print(f"Pass step 1: {nb_pass_step1['10dB']}")
    print(f"Pass step 2: {nb_pass_step2['10dB']}")
    print("")

    print("15dB")
    print("------")
    print(f"Pass step 1: {nb_pass_step1['15dB']}")
    print(f"Pass step 2: {nb_pass_step2['15dB']}")
```

## 1.1. Sorted by dB

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(2,2, figsize=(15,10))

f.suptitle("FRs detection")

ax[0,0].bar(x=["w1.1","w1.2","w2","w5","w10"],height=step1_db0)
ax[0,0].set_title("0 dB")
ax[0,0].grid()
ax[0,0].set_ylim(50,1500)
ax[0,1].bar(x=["w1.1","w1.2","w2","w5","w10"],height=step1_db5)
ax[0,1].set_title("5 dB")
ax[0,1].grid()
ax[0,1].set_ylim(50,1500)
ax[1,0].bar(x=["w1.1","w1.2","w2","w5","w10"],height=step1_db10)
ax[1,0].set_title("10 dB")
ax[1,0].grid()
ax[1,0].set_ylim(50,1500)
ax[1,1].bar(x=["w1.1","w1.2","w2","w5","w10"],height=step1_db15)
ax[1,1].set_title("15 dB")
ax[1,1].grid()
ax[1,1].set_ylim(50,1500)

plt.tight_layout()
```
